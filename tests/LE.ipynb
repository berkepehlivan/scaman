```python
# Generate a random dataset
from serial.le import LE
import numpy as np
import matplotlib.pyplot as plt
data = np.random.rand(100, 10)
colors = np.random.randint(0, 10, 100)

#from scipy.io import mmread

#data = mmread('data/breasttissue_10NN.mtx')
#data = data.tocsr().toarray()
#colors = mmread('data/breasttissue_10NN_label.mtx')

```

```python
## Creating test with numpy solver
# Create an instance of the LE class
le = LE(data, n_components=6, k=7, solver='numpy',sigma= 1,normalize=False)
embedding_numpy = le.fit_transform()
## plot 2d and 3d
le.plot_2d(embedding_numpy,labels=colors)
le.plot_3d(embedding_numpy,labels=colors)
```

```python
## Creating test with scipy solver
# Create an instance of the LE class
le = LE(data, n_components=6, k=7, solver='scipy',sigma= 1,normalize=False)
embedding_scipy = le.fit_transform()
## plot 2d and 3d
le.plot_2d(embedding_scipy,labels=colors)
le.plot_3d(embedding_scipy,labels=colors)
```

```python
## Creating test with slepc solver
# Create an instance of the LE class
le = LE(data, n_components=6, k=7, solver='slepc',slepc_solver='KRYLOVSCHUR',sigma= 1,normalize=True)
embedding_slepc = le.fit_transform()
## plot 2d and 3d
le.plot_2d(embedding_slepc,labels=colors)
le.plot_3d(embedding_slepc,labels=colors)
```
