```python
## Importing the breasttissue.mtx dataset
from scipy.io import mmread
import numpy as np

X = mmread('/Users/berke/Codes/scaman/tests/data/breasttissue_10NN.mtx').toarray()
color = mmread('/Users/berke/Codes/scaman/tests/data/breasttissue_10NN_label.mtx')
```

```python
## Importing the MDS class
from serial.mds import MDS
mds = MDS(n_components=3, eig_method='slepc',slepc_solver='ARNOLDI')
X_real = X.astype(float) # Convert X to a real data type
eigenvectors = mds.fit_transform(X_real)
#color_i = range(len(eigenvectors)) 
mds.plot(X_real, eigenvectors, color)
```

```python
## Importing the MDS class for scipy
from serial.mds import MDS
mds = MDS(n_components=4, eig_method='scipy')
eigenvectors = mds.fit_transform(X)
mds.plot(X, eigenvectors, color)
```

```python
## Importing the MDS class
from serial.mds import MDS
mds = MDS(n_components=2, eig_method='numpy')
eigenvectors = mds.fit_transform(X)
mds.plot(X, eigenvectors, color)
```
