```python
from serial.pca import PCA
pca = PCA(n_components=2, eig_method='numpy')
X_transformed = pca.fit_transform(X)
pca.plot(X_transformed, color=color, output_dir='output')
```

```python
## Applying PCA on Swissroll dataset
pca = PCA(n_components=2, eig_method='scipy')
X_transformed = pca.fit_transform(X)
pca.plot(X_transformed, color=color, output_dir='output')
```
