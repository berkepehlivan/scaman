```python
from eigensolvers.feast_eigensolver import FeastEigensolver
import numpy as np
np.random.seed(0)
M = np.random.rand(100,100)
# we make the random matrix symmetric, we do not need to do this for the laplacian!!!
symm = M@M.T
ei,ev=np.linalg.eig(symm) # numpy eigensolver
M = np.random.rand(100,30)
sei=np.sort(ei)
B=np.identity(100)
aei=sei[0:22] # for comparison with the result of the feast eigensolver 
result = FeastEigensolver(symm, B, 23, M, 4, 0, 1, 0.0000001, 1, 2,aei)
```

```python
# convert result.ei list to numpy array
result.ei=np.array(result.ei)
#check aei and result.ei are the same without allclose
np.allclose(aei,result.ei,atol=1e-3)
```
